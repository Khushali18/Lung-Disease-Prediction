This image dataset has 3 Lung X-ray classes namely Lung Opacity, Normal and Viral Pneumonia. Here in this pipeline we are preprocessing these three classes using CNN model.

### 1. Import Libraries

```python
import os
import warnings
warnings.filterwarnings("ignore")

from PIL import Image
import numpy as np
import pandas as pd
from Utilities import load_and_analyse_images, grayscale_and_resize, normalize_and_brighten
```

```python
original_folders = ['../Data/Lung X-ray Image/Lung Opacity', '../Data/Lung X-ray Image/Normal' , '../Data/Lung X-ray Image/Viral Pneumonia']
path_to_new_folder = '../Data/processed Lung X-ray Image/'
processed_folders = ['../Data/processed Lung X-ray Image/Lung Opacity', '../Data/processed Lung X-ray Image/Normal' , '../Data/processed Lung X-ray Image/Viral Pneumonia']
```
